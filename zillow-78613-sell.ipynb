# Houses for sell for 78613

```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import re
import time
import numpy as np

```

```python

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36'}
def getAndParseURL(url):
    result = requests.get(url,headers=headers, timeout=5)
    soup = BeautifulSoup(result.text, 'html.parser')
    return(soup)


base="https://www.zillow.com"
urlsell="https://www.zillow.com/homes/78613_rb/"
        
allapages=[urlsell]
index=1
a=getAndParseURL(urlsell)

while a.find("li",{"class":"zsg-pagination-next"}):
    delays = [7, 4, 6, 2, 10]
    delay = np.random.choice(delays)
    time.sleep(delay)
    nextpage=base+a.find("li",{"class":"zsg-pagination-next"}).a["href"]
    allapages.append(nextpage)
    a=getAndParseURL(nextpage)
```

```python
allapages
```

```python
allsearchpagelinks=[]
for k in allapages:
    delays = [4, 6, 2, 10]
    delay = np.random.choice(delays)
    time.sleep(delay)

    a=getAndParseURL(k)
    temp=a.find("div",{"id":"list-results"}).findAll("a",{"class":"zsg-photo-card-overlay-link routable hdp-link routable mask hdp-link"})
    allsearchpagelinks.extend(base+i["href"] for i in temp if i["href"].startswith("/homedetails") )
```

```python
beds=[]
baths=[]
sqfts=[]
estimateprices=[]
housetypes=[]
YearBuilts=[]
Heatings=[]
Coolings=[]
parkings=[]
floors=[]
lotsizes=[]
names=[]
numberOfRoomss=[]
streetAddresss=[]
addressLocalitys=[]
addressRegions=[]
postalCodes=[]
latitudes=[]
longitudes=[]
houseurls=[]   


for k in allsearchpagelinks:

    delays = [1,2,3,4,5,6,7,8,9]
    delay = np.random.choice(delays)
    time.sleep(delay)
    a=getAndParseURL(k)
    try:
        if a.find("span",{"class":"ds-status-details"}).text =='For sale':

            info=a.find("h3",{"class":"ds-bed-bath-living-area-container"}).findAll("span",{"class":"ds-bed-bath-living-area"})
            info=[i.span.text for i in info]

            try:
                #beds
                bed=int(info[0])
                beds.append(bed)
            except:
                beds.append(None)

            try:   
                #bath
                bath=float(info[1])
                baths.append(bath)
            except:
                baths.append(None)


            try:
                #sqrt
                sqrt=info[2]
                sqft=int(("").join((info[2]).split(",")))
                sqfts.append(sqft)
            except:
                sqfts.append(None)        



            try:
                #estimateprice
                estimateprice=a.find("span",{"class":"ds-value"}).text
                estimateprice=int(("").join(re.split(",|\$",estimateprice)))
                estimateprices.append(estimateprice)
            except:
                estimateprices.append(None)


            ##revised version

            try:
                fact=a.find("ul",{"class":"ds-home-fact-list"}).findAll("li",{"class":"ds-home-fact-list-item"})
                factvalue=[i.find("span",{"class":"ds-body ds-home-fact-value"}).text for i in fact]
                factlabel=[i.find("span",{"class":"ds-standard-label ds-home-fact-label"}).text for i in fact]
                factdict=dict(zip(factlabel,factvalue))

            except: 
                print(k)
                housetypes.append(None)
                YearBuilts.append(None)
                Heatings.append(None)
                Coolings.append(None)
                parkings.append(None)

            try:    
                #housetype
                housetype=factdict['Type:']
                housetypes.append(housetype)
            except: 
                housetypes.append(None)

            try:
                #'Year Built'
                YearBuilt=int(factdict['Year built:'])
                YearBuilts.append(YearBuilt)
            except: 
                YearBuilts.append(None)


            try:
                #Heating
                Heating=factdict['Heating:']
                Heatings.append(Heating)
            except: 
                Heatings.append(None)

            try:
                #'Cooling'
                Cooling=factdict['Cooling:']
                Coolings.append(Cooling)
            except: 
                Coolings.append(None)

            try:
                #parking
                parking=(("").join(re.split(" spaces|space",factdict['Parking:'])))
                parkings.append(parking)
            except: 
                parkings.append(None)


            try:
                #floor
                flr=int(a.find(text="Property").parent.parent.span.text)
                floors.append(flr)
            except:
                floors.append(None)

            #lotsize
            try:
                lotsize=int(("").join(re.split(" sqft|,|\.| acres",factdict['Lot:'])))
                lotsizes.append(lotsize)
            except:
                lotsizes.append(None)

            s=a.find("script",{"type":"application/ld+json"}).text

            name=json.loads(s)['name']#1414 Ridgewood Dr, Leander, TX 78641
            names.append(name)

            numberOfRooms=json.loads(s)['numberOfRooms']#3
            numberOfRoomss.append(numberOfRooms)

            streetAddress=json.loads(s)['address']["streetAddress"]#1414 Ridgewood Dr
            streetAddresss.append(streetAddress)

            addressLocality=json.loads(s)['address']['addressLocality']#Leander
            addressLocalitys.append(addressLocality)

            addressRegion=json.loads(s)['address']['addressRegion']#TX
            addressRegions.append(addressRegion)

            postalCode=json.loads(s)['address']['postalCode']#'78641'
            postalCodes.append(postalCode)

            latitude=json.loads(s)["geo"]['latitude']
            latitudes.append(latitude)

            longitude=json.loads(s)["geo"]['longitude']
            longitudes.append(longitude)

            houseurl=base+json.loads(s)["url"]
            houseurls.append(houseurl)

        else:
            continue
    except:
        continue
```

```python
df = pd.DataFrame({"beds":beds,
                   "baths":baths,
                   "sqft":sqfts,
                    "estimateprice":estimateprices,
                    "housetype":housetypes,
                    "YearBuilt":YearBuilts,
                    "Heating":Heatings,
                    "Cooling":Coolings,
                    "parking":parkings,
                    "lotsize":lotsizes,
                    "name":names,
                    "numberOfRooms":numberOfRoomss,
                    "streetAddress":streetAddresss,
                    "addressLocality":addressLocalitys,
                    "addressRegion":addressRegions,
                    "postalCode":postalCodes,
                    "latitude":latitudes,
                    "longitude":longitudes,
                    "houseurl":houseurls

  })

df.to_csv('zillow-78613-sell.csv')
```
